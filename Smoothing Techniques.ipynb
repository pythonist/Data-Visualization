```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('Closing Stock Prices.xlsx',skiprows = 2)
```

```python
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='IBM',ax=ax)
plt.legend(['IBM'])
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("IBM stock irregularities")
plt.show()
plt.figure(figsize=(5,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='INTC',ax=ax)
plt.legend(['INTC'])
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("INTC stock irregularities")
plt.show()
plt.figure(figsize=(5,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='CSCO',ax=ax)
plt.legend(['CSCO'])
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("CSCO stock irregularities")
plt.show()
plt.figure(figsize=(5,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='GE',ax=ax)
plt.legend(['GE'])
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("GE stock irregularities")
plt.show()
plt.figure(figsize=(5,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='DJ Industrials \nIndex',ax=ax)
plt.legend(['DJ Industrials \nIndex'])
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("DJ Industrials \nIndex stock irregularities")
plt.show()
```

# MOVING AVERAGE

```python
df["IBM_MA"]=df["IBM"].rolling(window=3).mean().shift(1)
df["GE_MA"]=df["GE"].rolling(window=3).mean().shift(1)
df["INTC_MA"]=df["INTC"].rolling(window=3).mean().shift(1)
df["CSCO_MA"]=df["CSCO"].rolling(window=3).mean().shift(1)
df["DJ Industrials \nIndex_MA"]=df["DJ Industrials \nIndex"].rolling(window=3).mean().shift(1)
```

```python
plt.figure(figsize=(10,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='IBM',ax=ax)
df.plot(kind='line',x='Date',y='IBM_MA', color='red', ax=ax)
plt.legend(['IBM','IBM_MA'])
plt.xlabel("Date")
plt.ylabel("Stock Prices")
plt.title("IBM after moving average")
plt.show()
plt.figure(figsize=(10,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='INTC',ax=ax)
df.plot(kind='line',x='Date',y='INTC_MA', color='red', ax=ax)
plt.legend(['INTC','INTC_MA'])
plt.xlabel("Date")
plt.ylabel("Stock Prices")
plt.title("INTC after moving average")
plt.show()
plt.figure(figsize=(10,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='CSCO',ax=ax)
df.plot(kind='line',x='Date',y='CSCO_MA', color='red', ax=ax)
plt.legend(['CSCO','CSCO_MA'])
plt.xlabel("Date")
plt.ylabel("Stock Prices")
plt.title("CSCO after moving average")
plt.show()
plt.figure(figsize=(10,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='GE',ax=ax)
df.plot(kind='line',x='Date',y='GE_MA', color='red', ax=ax)
plt.legend(['GE','GE_MA'])
plt.xlabel("Date")
plt.ylabel("Stock Prices")
plt.title("GE after moving average")
plt.show()
plt.figure(figsize=(10,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='DJ Industrials \nIndex',ax=ax)
df.plot(kind='line',x='Date',y='DJ Industrials \nIndex_MA', color='red', ax=ax)
plt.legend(['DJ Industrials \nIndex','DJ Industrials \nIndex_MA'])
plt.xlabel("Date")
plt.ylabel("Stock Prices")
plt.title("DJ Industrials \nIndex after moving average")
plt.show()
```

# EXPONENTIAL SMOOTHING

```python
df["IBM_ES"]=df["IBM"].ewm(alpha=0.3).mean()
df["INTC_ES"]=df["INTC"].ewm(alpha=0.3).mean()
df["CSCO_ES"]=df["CSCO"].ewm(alpha=0.3).mean()
df["GE_ES"]=df["GE"].ewm(alpha=0.3).mean()
df["DJ Industrials \nIndex_ES"]=df["DJ Industrials \nIndex"].ewm(alpha=0.3).mean()
```

```python
plt.figure(figsize=(12,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='IBM',ax=ax)
df.plot(kind='line',x='Date',y='IBM_MA', color='red', ax=ax)
df.plot(kind='line',x='Date',y='IBM_ES', color='green', ax=ax)
plt.xlabel("Date")
plt.legend(['Deafult','using MA','using ES'])
plt.ylabel("Stock Price for IBM")
plt.title("after moving average and smoothing")
plt.show()
plt.figure(figsize=(12,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='INTC',ax=ax)
df.plot(kind='line',x='Date',y='INTC_MA', color='red', ax=ax)
df.plot(kind='line',x='Date',y='INTC_ES', color='green', ax=ax)
plt.xlabel("Date")
plt.legend(['Deafult','using MA','using ES'])
plt.ylabel("Stock Price for INTC")
plt.title("after moving average and smoothing")
plt.show()
plt.figure(figsize=(12,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='CSCO',ax=ax)
df.plot(kind='line',x='Date',y='CSCO_MA', color='red', ax=ax)
df.plot(kind='line',x='Date',y='CSCO_ES', color='green', ax=ax)
plt.xlabel("Date")
plt.legend(['Deafult','using MA','using ES'])
plt.ylabel("Stock Price for CSCO")
plt.title("after moving average and smoothing")
plt.show()
plt.figure(figsize=(10,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='GE',ax=ax)
df.plot(kind='line',x='Date',y='GE_MA', color='red', ax=ax)
df.plot(kind='line',x='Date',y='GE_ES', color='green', ax=ax)
plt.xlabel("Date")
plt.legend(['Deafult','using MA','using ES'])
plt.ylabel("Stock Price for GE")
plt.title("after moving average and smoothing")
plt.show()
plt.figure(figsize=(10,5))
ax=plt.gca()
df.plot(kind='line',x='Date',y='DJ Industrials \nIndex',ax=ax)
df.plot(kind='line',x='Date',y='DJ Industrials \nIndex_MA', color='red', ax=ax)
df.plot(kind='line',x='Date',y='DJ Industrials \nIndex_ES', color='green', ax=ax)
plt.xlabel("Date")
plt.legend(['Deafult','using MA','using ES'])
plt.ylabel("Stock Price for DJ Industrials \nIndex")
plt.title("after moving average and smoothing")
plt.show()
```

# ERROR METRICS

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
df=df.dropna(how='any')
print('RMSE for Moving Average:',mean_squared_error(df["IBM"], df["IBM_MA"], squared=False))
print('MAD for Moving Average:',mean_absolute_error(df["IBM"], df["IBM_MA"]))
print('RMSE for Exp Smoothing:',mean_squared_error(df["IBM"], df["IBM_ES"], squared=False))
print('MAD for Exp Smoothing:',mean_absolute_error(df["IBM"], df["IBM_ES"]))
```

```python
print('RMSE for Moving Average:',mean_squared_error(df["INTC"], df["INTC_MA"], squared=False))
print('MAD for Moving Average:',mean_absolute_error(df["INTC"], df["INTC_MA"]))
print('RMSE for Exp Smoothing:',mean_squared_error(df["INTC"], df["INTC_ES"], squared=False))
print('MAD for Exp Smoothing:',mean_absolute_error(df["INTC"], df["INTC_ES"]))
```

```python
print('RMSE for Moving Average:',mean_squared_error(df["CSCO"], df["CSCO_MA"], squared=False))
print('MAD for Moving Average:',mean_absolute_error(df["CSCO"], df["CSCO_MA"]))
print('RMSE for Exp Smoothing:',mean_squared_error(df["CSCO"], df["CSCO_ES"], squared=False))
print('MAD for Exp Smoothing:',mean_absolute_error(df["CSCO"], df["CSCO_ES"]))
```

```python
print('RMSE for Moving Average:',mean_squared_error(df["GE"], df["GE_MA"], squared=False))
print('MAD for Moving Average:',mean_absolute_error(df["GE"], df["GE_MA"]))
print('RMSE for Exp Smoothing:',mean_squared_error(df["GE"], df["GE_ES"], squared=False))
print('MAD for Exp Smoothing:',mean_absolute_error(df["GE"], df["GE_ES"]))
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import statsmodels.api as sm
%matplotlib inline
```

```python
plt.figure(figsize=(15,5))
plt.title('Electricity production per year')
plt.xlabel('Year')
plt.legend(['US','Canada','France'])
plt.ylabel('Production in Kwh')
plt.scatter(df1['Year'],df1['US'],color='red',marker='o')
plt.scatter(df1['Year'],df1["Canada"],color='Blue',marker='x')
plt.scatter(df1['Year'],df1["France"],color='green',marker='+')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(df1[['Year']].values,df1[['US']])
print('For 2007:',reg.predict([[2007]]))
print('For 2008:',reg.predict([[2008]]))
print('For 2009:',reg.predict([[2009]]))
print(reg.coef_,reg.intercept_)
plt.plot(df1['Year'],reg.predict(df1[['US']]),color='blue')
plt.title("Forecasting for US")
```
